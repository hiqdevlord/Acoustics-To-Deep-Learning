But why do our position-velocity and velocity-acceleration graphs form circles? What could a circle possible have to do with a mass on a spring moving back and forth? And why is this video called "waffles and harmonic motion"??
 


Well, first of all, waffles are delicious, and secondly, circularly shaped. When an object moves in a circle, its motion is constrained in space in a very specific way. Circles seem like very natural shapes to us, but if we consider Newton’s first law, which says that objects really prefer to move in straight lines, then circles are kind of a weird way to move. In fact, keeping an object moving in circular path requires the object to be constantly accelerated towards the center of the circle. We give this acceleration a special name - centripetal. 

Now that we know something about the position and acceleration of objects moving in circular paths, let's compare this motion to the motion of our spring. Since our spring only moves in 1 spatial dimension, we would like to know how object moving around a circles move in 1 spatial dimension. Circles are, by definition, 2-dimensional, so we need to "project" our motion down to 1 dimension. One way to do this physically is to cast a shadow of the object in motion - the shadow moves in one less dimension than the object does. 

Mathematically, we can compute projected motions using right triangles. As our object moves around the circle, for each point on the circle we can make a large right triangle who's bottom leg is x long, and radius is r long. Since our centripetal acceleration points straight towards our center, we can draw a second right triangle whos hypotenuse is ac long, and who's bottom leg corresponds to the x-component of our centripetal acceleration. 

Since our triangles are similar, we see that the base legs, x, and a x, must be proportional. We can express this in a ratio, and if we arrange our equation, we see that our position x is proportional to some constants time minus acceleration! Remember that our spring yielded almost an identical formula, the only difference is our constants. 

Since constants only change the rate and scale of motion and not the motion itself, we've found our connection! Moving in a circle requires a special relationship between position and acceleration, which is the same exact relationship that a spring imposes. These relationships govern the motion of each object, and since they're the same, the objects will follow the same paths. 

In the circle case we started by assuming a motion and figured out the acceleration created by the motion. In the spring case we assumed an acceleration and figured out the motion. In both cases we found the same exact relationship. Position is proportional to negative acceleration. This is the connection. Springs move like circles and circles accelerate like springs.
 


Now that we see the concepts clearly, there is some more math we should really understand here to tie things up nicely. Depending on how thoroughly you learned trigonometry, you already know that we have special names for the functions that results from projecting circles into a single dimension. 
 


The official names for these functions are sine and cosine. Cosine gives the x value as we move around a unit circle, while sine gives the y. In calculus class we learn that sine and cosine have some special properties when differentiated. The derivate, or slope, or a sin wave is cosine. And the derivate of cosine is negative sin. So if we differentiate sin or cosine twice, we get the minus version of what we started with. 

If we look back at our equation of motion for a spring, we can rewrite acceleration as the second derivative of position, yielding a differential equation. The solution to the equation must give the minus version of itself when differentiated twice, which is, of course, exactly what sin and cosine do. 

If we go back to our numerical analysis from part 1, where we just computed the motion of our spring step by step, we can compare our resulting positions to a cosine function and see that they match to the third decimal place!

In our spring analysis we assumed that spring constant k and mass m were one, so let's add these back in. Since different springs oscillate at different rates, we'll add a frequency term, omega 0, to our cos solutions to capture this. We'll call omega 0 the "natural frequency" of the oscillator. When we include our constants and works things out, we see that the natural frequency of our oscillator will be equal to the root of k over m.  So given the stiffness of the spring and our mass we can compute our frequency of oscillation. 


$$
kx = -ma
$$

$$
kx = -m\frac{d^2x}{dt^2}
$$

$$
\frac{d^2x}{dt^2} = -\frac{k}{m}x
$$

Finally, our spring and mass could be started moving at various speeds and positions, so our solution should take this into account. We could add a "phase term" inside our cosine to allow our graph to start at various points, but a simpler approach is to take advantage of the fact that we can express an arbitrary phase as a sum of sines and cosines, and write our solution as A cos omega 0 t + B sin omega 0 t. 

```python
def plotCosWithPhase(delta):
    w = 1
    
    t = arange(0,10, 0.1)
    x = cos(w*t+delta)
    
    A = cos(delta)
    B = -sin(delta)
    
    xA = A*cos(w*t)
    xB = B*sin(w*t)
    
    fig = figure(0, (8, 4))
    
    plot(t, x, 'x')
    grid(1)
    
    plot(t,xA)
    plot(t,xB)
    plot(t, xA+xB)
    
    legend(['Original', 'A Component', 'B Component', 'Sum'])
```

```python
interact(plotCosWithPhase, delta=(-2,2,0.1))
```

This is our general solution to or spring and mass problem. Using techniques from differential equations, we can show that this is the unique, or only solution to our system. Given an initial position and velocity, we can solve for A and B in to determine how our oscillator will move in specific problems.
